```python
selectPsmstr = ("select Ps_par as 父件, Ps_comp as	子件, pt_desc1 as 中文說明1, pt_desc2 as 中文說明2, xen_desc1 as 英文說明1, xen_desc2 as 英文說明2, pt_group as 組別,Ps_op as 作業碼, Ps_qty_per as	單位數量 from ps_mstr Join pt_mstr On pt_part = Ps_comp left join xen_mstr on ps_comp = xen_part where (Ps_start <= GETDATE() or Ps_start is null) and (Ps_end >= GETDATE() or Ps_end is null)")
cursor.execute(selectPsmstr)
result=cursor.fetchall()
psmstr = pd.DataFrame(result)
```

```python
selectWod = ("Select Wod_part as 子件,  pt_desc1 as 中文說明1, pt_desc2 as 中文說明2, xen_desc1 as 英文說明1, xen_desc2 as 英文說明2, Pt_mstr.Pt_group as 組別,  Wod_qty_iss as 發料量, wod_code as 機號 from wod_det join pt_mstr on wod_part = pt_mstr.pt_part left join xen_mstr on Wod_part = xen_part")
cursor.execute(selectWod)
result=cursor.fetchall()
wod = pd.DataFrame(result)
```

```python
selectPTmstr = ("select pt_part as 子件, pt_desc1 as 中文說明1, pt_desc2 as 中文說明2 , xen_desc1 as 英文說明1, xen_desc2 as 英文說明2, pt_group as 組別 from pt_mstr left join xen_mstr on pt_part = xen_part")
cursor.execute(selectPTmstr)
result=cursor.fetchall()
PTmstr = pd.DataFrame(result)
```

```python

rule = pd.read_csv(rulefile)
rule = rule.drop(columns=["研發更改V2","研發更改V2進度","研發更改V1","研發更改V1.1"])
rule['中文說明1'],rule['中文說明2'],rule['英文說明1'],rule['英文說明2'] = None,None,None,None
rule
```

```python
error = []
# 根據PTmstr更新rule中每個零件的中英文說明
for index, Row in rule.iterrows():
    if len(PTmstr[PTmstr['子件']==Row['零件']])>0:
        rule.at[index,"中文說明1"] = PTmstr[PTmstr['子件']==Row['零件']].iloc[0]['中文說明1']
        rule.at[index,"中文說明2"] = PTmstr[PTmstr['子件']==Row['零件']].iloc[0]['中文說明2']
        rule.at[index,"英文說明1"] = PTmstr[PTmstr['子件']==Row['零件']].iloc[0]['英文說明1']
        rule.at[index,"英文說明2"] = PTmstr[PTmstr['子件']==Row['零件']].iloc[0]['英文說明2']
    else:
        error.append(Row['零件']) # 檢查零件包內錯誤料號
        print(Row['零件'],"不存在") 
```
