```python
import pandas as pd
import numpy as np
nan = np.nan
import warnings
warnings.filterwarnings("ignore")
```

```python
a = pd.read_csv("rule.csv" )
a
```

```python
for i in range(len(a)):
    if a["零件包料號"][i] is nan:
        a["零件包料號"][i] = a["零件包料號"][i-1]
    if a["名稱"][i] is nan:
        a["名稱"][i] = a["名稱"][i-1]
```

```python
def transcode1(a):
    spt = a['代號'].split('|')
    if a['計算方式'] == "同車台工單數量":
        return "車台工單"
    if spt[0] == "0":
        return "固定"
    if spt[0] == "f":
        return "車台口數+2"
    
    if len(spt[0].split(","))>1:
        return ' or '.join(spt[0].split(","))
    else:
        return spt[0]
def transcode2(a):
    spt = a['代號'].split('|')
    if (len(spt)>1):
        return spt[1]
    else:
        if a['參照對象'] == "固定":
            return a["量"]
        else:
            return 1
def classref(a):
    spt = a["參照對象"].split(",")
    if spt[0] in ["車台工單", "固定"]:
        return spt[0]
    if spt[0] == "車台口數+2":
        return "車台口數"
    if len(spt)>1:
        return "2組別擇一"
    elif len(spt[0])>4:
        return "料號"
    else:
        return "組別"
    
a['參照對象'] = a.apply(transcode1,axis=1)
a['參照倍數'] = a.apply(transcode2,axis=1)
a[a["參照對象"]=="車台工單"]
# a = a.drop(columns=["計算方式","代號"],axis=1)
# a = a.drop(columns=["代號"],axis=1)
a["參照類別"] = a.apply(classref,axis=1)
```
